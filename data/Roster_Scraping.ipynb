```python
import requests
import bs4
import pandas as pd
```

# Male Baseball

```python
# This function work for male baseball, basketball, (some)soccer, volleyball, waterpolo
def oneYear_roster(roster_url):
    response = requests.get(roster_url)
    soup = bs4.BeautifulSoup(response.text)
    athletes = soup.find_all('tbody')[2] # found after checking the html source code
    data = {'JerseyNum': [],
            'Name': [],
            'Position': [],
            'Height': [],
            'Weight':[],
            'Major': [],
            'Class':[],
            'HomeTownHighSchool':[]}
    for athlete in athletes.find_all('tr'):
        try:
            data['JerseyNum'].append(athlete.find('td', attrs={'class': 'roster_jerseynum'}).text)
        except:
            data['JerseyNum'].append("")
        try:
            data['Name'].append(athlete.find('td', attrs={'class': 'sidearm-table-player-name'}).find('a').text)
        except: 
            data['Name'].append("")
        try:
            data['Position'].append(athlete.find('td', attrs={'class': 'rp_position_short'}).text)
        except:
            data['Position'].append("")
        try:
            data['Height'].append(athlete.find('td', attrs={'class': 'height'}).text)
        except:
            data['Height'].append("")
        try:
            data['Weight'].append(athlete.find('td', attrs={'class': 'rp_weight'}).text)
        except:
            data['Weight'].append("")
        try:
            data['Major'].append(athlete.find('td', attrs={'class': 'player_major'}).text)
        except:
            data['Major'].append("")
        data['Class'].append(athlete.find('td', attrs={'class': 'roster_class'}).text)
        data['HomeTownHighSchool'].append(athlete.find('td', attrs={'class': 'hometownhighschool'}).text)
    
    return pd.DataFrame(data)
```

```python
def male_baseball_rosters_all():
    baseball_rosters_all = {}
    for year in range(2006,2025): # just to note, no data for major before 2020
    #     print(year)
        baseball_url = f"https://ucsdtritons.com/sports/baseball/roster/{year}?print=true"
        baseball_rosters_all[str(year)] = oneYear_roster(baseball_url)
    
    dfs = None
    for year, df in baseball_rosters_all.items():
        df['Year'] = year
        if dfs is None:
            dfs = df
        else:
            dfs = pd.merge(dfs, df, how='outer')
    
    dfs = dfs[["Year","JerseyNum","Name","Position","Height","Weight","Major","Class","HomeTownHighSchool"]]
    return dfs
```

```python
dfs = male_baseball_rosters_all()
dfs.to_csv('male_baseball_rosters.csv', index=False) # save as csv
pd.read_csv('male_baseball_rosters.csv') # checking
```

# Fencing 
both Male and Female in the same pdf

```python
def fencing_oneYear_roster(roster_url, women = False):
    response = requests.get(roster_url)
    soup = bs4.BeautifulSoup(response.text)
    if women == True:
        athletes = soup.find_all('tbody')[3] # women
    else:
        athletes = soup.find_all('tbody')[2] # men
    data = {'JerseyNum': [],
            'Name': [],
            'Position': [],
            'Height': [],
            'Weight':[],
            'Major': [],
            'Class':[],
            'HomeTownHighSchool':[]}
    for athlete in athletes.find_all('tr'):
        try:
            data['JerseyNum'].append(athlete.find('td', attrs={'class': 'roster_jerseynum'}).text)
        except:
            data['JerseyNum'].append("")
        data['Name'].append(athlete.find('td', attrs={'class': 'sidearm-table-player-name'}).find('a').text)
        data['Position'].append(athlete.find('td', attrs={'class': 'rp_position_short'}).text)
        try:
            data['Height'].append(athlete.find('td', attrs={'class': 'height'}).text)
        except:
            data['Height'].append("")
        try:
            data['Weight'].append(athlete.find('td', attrs={'class': 'rp_weight'}).text)
        except:
            data['Weight'].append("")
        try:
            data['Major'].append(athlete.find('td', attrs={'class': 'player_major'}).text)
        except:
            data['Major'].append("")
        data['Class'].append(athlete.find('td', attrs={'class': 'roster_class'}).text)
        data['HomeTownHighSchool'].append(athlete.find('td', attrs={'class': 'hometownhighschool'}).text)
    
    return pd.DataFrame(data)
```

```python
# url format https://ucsdtritons.com/sports/fencing/roster?print=true

def fencing_rosters_all(women=False):
    fencing_rosters_all = {}
    for year in range(2005,2024): 
#         print(year)
        year_str = str(year)+"-"+ str(year+1)[-2:]
        fencing_url = f"https://ucsdtritons.com/sports/fencing/roster/{year_str}?print=true"
        fencing_rosters_all[str(year)] = fencing_oneYear_roster(fencing_url,women)

    dfs = None
    for year, df in fencing_rosters_all.items():
        df['Year'] = year
        if dfs is None:
            dfs = df
        else:
            dfs = pd.merge(dfs, df, how='outer')
    
    dfs = dfs[["Year","JerseyNum","Name","Position","Height","Weight","Major","Class","HomeTownHighSchool"]]
    return dfs
```
